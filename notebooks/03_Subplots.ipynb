```python
fig = ep.PlotlyFigure('../templates/subplots.yml', title='Subplot Example', rows=2, cols=2, subplot_titles=['Plot 1', 'Plot 2', 'Plot 3', 'Plot 4'])

fig.add_scatter([2, 2, 3, 6, 10], [1, 2, 4, 10, 12], 'Type I', ['a', 'b', 'c', 'd', 'e'], 0, row=1, col=1)
fig.add_scatter([2, 3, 10, 12, 20], [2, 9, 8, 7, 8], 'Type II', ['a', 'b', 'c', 'd', 'e'], 1, row=1, col=2)
fig.add_scatter([3, 7, 8, 6, 15], [10, 12, 32, 22, 40], 'Type III', ['a', 'b', 'c', 'd', 'e'], 2, row=2, col=1)
fig.add_scatter([2, 6, 7, 5, 14], [9, 11, 31, 21, 39], 'Type IV', ['a', 'b', 'c', 'd', 'e'], 3, row=2, col=2)
fig.add_scatter([4, 8, 9, 7, 16], [11, 13, 31, 23, 39], 'Type V', ['a', 'b', 'c', 'd', 'e'], 4, row=1, col=1)
fig.to_png(show=True)
```
