```python
import easy_plotly as ep
```

```python
import networkx as nx
import math
```

```python
G = nx.petersen_graph()

# set 2D positions

for i in range(5):
    theta = 2 * math.pi * i / 5 + math.pi / 2
    G.nodes[i]['pos'] = (2 * math.cos(theta), 2 * math.sin(theta))  # outer vertices
    G.nodes[5 + i]['pos'] = (math.cos(theta), math.sin(theta))      # inner vertices

# print(nx.get_node_attributes(G, 'pos'))
```

```python
ep.PlotlyGraph(G, '../templates/graph.yml').to_png(show=True)
```

```python
ep.PlotlyGraph(
    G,
    '../templates/graph.yml',
    node_text_position='top center',
    node_size=40,
    node_color=int,                   # color vertices based on their IDs
    node_text_font_size={4: 30},
    node_border_width=4,
    node_opacity=0.8,
    edge_color={(0, 1): '#ff0000'},
    edge_width={(1, 2): 10},
).to_png(show=True)
```

```python
r = math.sqrt(3)
def mid(a, b):
    return tuple((a[i] + b[i]) / 2 for i in range(3))

pos3d = {
    0: (0, 0, r),
    3: (1, 1/r, 0),
    6: (-1, 1/r, 0),
    7: (0, -2/r, 0),
}
pos3d[1] = mid(pos3d[0], pos3d[6])
pos3d[2] = mid(pos3d[3], pos3d[7])
pos3d[4] = mid(pos3d[0], pos3d[3])
pos3d[5] = mid(pos3d[0], pos3d[7])
pos3d[8] = mid(pos3d[3], pos3d[6])
pos3d[9] = mid(pos3d[6], pos3d[7])

for i in range(10):
    G.nodes[i]['pos'] = pos3d[i]
```

```python
ep.PlotlyGraph(G, '../templates/graph.yml', node_color=int).to_png(show=True)
```

## Arrows and edge text

```python
G = nx.petersen_graph()

# set 2D positions

for i in range(5):
    theta = 2 * math.pi * i / 5 + math.pi / 2
    G.nodes[i]['pos'] = (2 * math.cos(theta), 2 * math.sin(theta))  # outer vertices
    G.nodes[5 + i]['pos'] = (math.cos(theta), math.sin(theta))      # inner vertices
```

```python
edge_text = {
    (0, 4): 123,
    (3, 4): 456,
    (6, 9): 'edge text',
    (2, 3): 789,
    (1, 2): 10,
}

ep.PlotlyGraph(
    G,
    '../templates/graph.yml',
    edge_text=edge_text,
    edge_color={(5, 7): 'green'},
    edge_arrow_show={(0, 1): True, (5, 7): True},
    edge_text_font_color={(6, 9): 'red'},
    edge_text_font_size={(6, 9): 20},
    edge_text_font_family={(6, 9): 'Geogia'},
    edge_text_offset={(2, 3): 'ccw(15)', (1, 2): (-10, 0)}
).to_png(show=True)
```
